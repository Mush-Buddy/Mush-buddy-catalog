```python
# mushroom class for easy fill and export
class Mushroom:
    
    def __init__(self, nameScientific, wikipage):
        
        self.nameScientific = nameScientific
        self.image_links = []
        self.nameCommon = ""
        self.description = ""
        self.foundInStates = []
        self.seasons = []
        self.hasVeil = False
        self.veilType = ""
        self.stemColors = []
        self.capShape = ""
        self.capColors = []
        self.hasGills = False
        self.gillType = ""
        self.gillAttatchmentStyle = ""
        self.sporePrint = []
        self.edibility = ""
        self.dateCreated = "2/22/22"
        self.license = ""
        self.wikiPage = wikipage
        self.citationList = []
        self.ecology = ""
        
    def addImageLink(self, link):
        self.image_links.append(link)
        
    def setCommonName(self, nameCommon):
        self.nameCommon = nameCommon
        
    def setDescription(self, description):
        self.description = description

    def addState(self, state):
        self.foundInStates.append(state);

    def addSeason(self, season):
        self.seasons.append(season)

    def setHasVeil(self, VeilBool):
        self.hasVeil=VeilBool
        
    def setVeilType(self, veilType):
        self.veilType = veilType
    
    def addStemColor(self, color):
        self.stemColors.append(color)
        
    def setCapShape(self, capShape):
        self.capShape = capShape

    def addCapColor(self, color):
        self.capColors.append(color)

    def setHasGills(self, gillBool):
        self.hasVeil=gillBool
    
    def setGillType(self, gillType):
        self.gillType = gillType
        
    def setGillAttachmentStyle(self, gillAttachment):
        self.gillAttatchmentStyle = gillAttachment
        
    def setEdibility(self, edibility):
        self.edibility = edibility
    
    def setEcology(self, ecology):
        self.ecology = ecology
        
    def addSporePrintColor(self, color):
        self.sporePrint.append(color)
    
    def addCitation(self, citation):
        self.citationList.append(citation)
        
    def addLicense(self, license):
        self.license = license
        
    def __str__(self):
        return str(self.nameScientific)
    
```

```python
def makeMushroom(page):
    

    print("\n\nNEW PAGE CRAWLING:\t", page, "\n\n")

    driver = webdriver.Chrome()
    driver.get(page)
    
    
    try:
        
        temp_scientificName = findScientificName(driver)
        temp_image = findImage(driver)
        temp_description = findDescription(driver)
        
        
        # make the mushroom    
        curr_mush = Mushroom(temp_scientificName, page)
        
        # adding the imaage
        curr_mush.addImageLink(temp_image)
        
        #add the description
        curr_mush.setDescription(temp_description)
        

        # now find the infobox
        infobox = findBox(driver)
        
        setMushroomProperties(curr_mush, infobox)
        
        driver.quit()
        return temp_scientificName, curr_mush

    except Exception as e: print(e)
            
    driver.quit()


def findImage(driver):
    plants = driver.find_element(By.TAG_NAME, "a img").get_attribute("src")
    return plants


def findScientificName(driver):
    temp_nameScientific = driver.find_element(By.ID, "firstHeading").text
    return temp_nameScientific

def findDescription(driver):
    ele = driver.find_elements_by_xpath("//span[contains(text(),'Description')]/parent::h2/following-sibling::p");
    #keep the first paragraph
    return (ele[0].text)
    

def findBox(driver):
#     description = driver.findElement(By.cssSelector("p:contains('__')"));
    plants = driver.find_elements(By.CLASS_NAME, "infobox")
    box = plants[1].text
    item_list = parser(box)
    return item_list
    
     
def parser(box):
    a = box.split("\n")
    key = a[0]
    useful = a[2:-1]         
    return useful

def setMushroomProperties(mushroom, infobox):
    
    if mushroom:
        for item in infobox:
            
            if "on hymenium" in item:
                mushroom.setHasVeil(True)
                temp_hymenium = item.split(' ')[0]
                mushroom.setVeilType(temp_hymenium)
                
            if "cap is" in item:
                temp_shape = item.split(' ')[2]
                mushroom.setCapShape(temp_shape)
                
            if "hymenium is" in item:
                temp_attach = item.split(' ')[2]
                mushroom.setGillAttachmentStyle(temp_attach)
                
            if "stipe has a" in item:
                mushroom.setHasVeil(True)
                temp_veil = item.split(' ')[3]
                mushroom.setVeilType(temp_veil)
                
            if "spore print is" in item:
                temp_spore = item.split(' ')[3]
                mushroom.addSporePrintColor(temp_spore)
                
            if "ecology is" in item:
                temp_ecology = item.split(' ')[2]
                mushroom.setEcology(temp_ecology)
    


wikiPages2crawl = ["https://en.wikipedia.org/wiki/Agaricus_bisporus"]
makeMushroom(wikiPages2crawl[0])
```

```python
def makeDictionary(mushroom_list):
    
    mushroom_dict = {}
    
    for item in mushroom_list:
        try:
            temp = makeMushroom(item)
            mushroom_dict[temp[0]]=temp[1]
        except:
            print("failed and making mushroom\n", item)
    return mushroom_dict
```
