```python
# web scraper with selenium following tech with tim tutorial
# Emilie Hopkinson
# 02/17/22

# dependencies
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def makeCSVfiles(pages2Crawl):
    
    for page in pages2Crawl:

        print("\n\nNEW PAGE CRAWLING:\t", page, "\n\n")
        
        driver = webdriver.Chrome()
        driver.get(page)
        try:
#             findImage(driver)
#             findScientificName(driver)
            findBox(driver)
#             findDescription(driver)
            
              
        
        except Exception as e: print(e)
            
    driver.quit()


def findImage(driver):
    plants = driver.find_element(By.TAG_NAME, "a img").get_attribute("src")
    print(plants, "images")


def findScientificName(driver):
    temp_nameScientific = driver.find_element(By.ID, "firstHeading").text
    print("success at creating scientific name:\t", temp_nameScientific)

def findDescription(driver):
    ele = driver.find_elements_by_xpath("//span[contains(text(),'Description')]/parent::h2/following-sibling::p");
    #keep the first paragraph
    print(ele[0].text)
    

def findBox(driver):
#     description = driver.findElement(By.cssSelector("p:contains('__')"));
    plants = driver.find_elements(By.CLASS_NAME, "infobox")
    box = plants[1].text
    parser(box)
     

def parser(box):
    a = box.split("\n")
    key = a[0]
    useful = a[2:-1]
    useless = ["a", "is", "has"]
    
    for text in useful:
        words = text.split(" ")
        for word in words:
            if word in useless:
                continue
             
            
    return
    


# wikiPages2crawl = ["https://en.wikipedia.org/wiki/Agaricus_bisporus", "https://en.wikipedia.org/wiki/Pleurotus_ostreatus", "https://en.wikipedia.org/wiki/Cantharellus"]
wikiPages2crawl = ["https://en.wikipedia.org/wiki/Agaricus_bisporus"]



makeCSVfiles(wikiPages2crawl)
```
